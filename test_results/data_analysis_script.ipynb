```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data from CSV files
RMQ_preprocess_times = pd.read_csv('RMQ_preprocess_times.csv')
RMQ_query_times = pd.read_csv('RMQ_query_times.csv')

# Plotting Preprocess Time vs. Sequence Length
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.plot(RMQ_preprocess_times['SequenceLength'], RMQ_preprocess_times['Time'], marker='o', color='#FF204E')
plt.title('Preprocessing Time vs. Sequence Length')
plt.xlabel('Sequence Length')
plt.ylabel('Time (microseconds)')
plt.grid(True)

# Plotting Query Time vs. Sequence Length
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.plot(RMQ_query_times['SequenceLength'], RMQ_query_times['Time'], marker='o', color='#A0153E')
plt.title('Average Query Time vs. Sequence Length')
plt.xlabel('Sequence Length')
plt.ylabel('Time (microseconds)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
